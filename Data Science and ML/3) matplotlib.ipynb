```python
max_temp=[40,50,60,32,51,75,45]
min_temp=[35,28,12,8,31,45,32]
avg_temp=[38,39,36,20,43,60,38]
week_days=[1,2,3,4,5,6,7]
```

# Bar Chat

```python
students =['ali','asim','umer','mohsin','ijaz','nasir']
mids =[60,53,83,70,91,21]
finals=[45,71,90,65,80,58]
```

```python
xpos=np.arange(len(students))
xpos
```

```python
plt.xticks(xpos,students)
plt.bar(xpos,mids,label='mids',color='green')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Stuent Report')
plt.legend(fontsize='large')
```

```python
plt.xticks(xpos,students)
plt.bar(xpos+0.2,mids,width=0.4,label='mids',color='green')
plt.bar(xpos-0.2,finals,width=0.4,label='finals',color='blue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Stuent Report')
plt.legend(fontsize='small')
plt.grid()
```

```python
plt.yticks(xpos,students)
plt.barh(xpos+0.2,mids,label='mids',color='green')
plt.barh(xpos-0.2,finals,label='finals',color='blue')
plt.ylabel('Students')
plt.xlabel('Marks')
plt.title('Stuent Report')
plt.legend(fontsize='small')
plt.grid()
```

# Histograms

```python
sugar_level_men=[90,101,145,83,65,77,93,98,114,122,200]
sugar_level_women=[61,85,104,77,119,125,160,133,88,110]
```

```python
plt.hist(sugar_level_men,bins=[80,100,125,200],rwidth=0.97,)
```

```python
plt.xlabel('Blood Pressure')
plt.ylabel('number of persons')
plt.title('Blood pressure survey report')
plt.hist([sugar_level_men,sugar_level_women],bins=[80,100,125,200],rwidth=0.97,label=['men','women'])
plt.legend()
```

```python

plt.xlabel('Blood Pressure')
plt.ylabel('number of persons')
plt.title('Blood pressure survey report')
plt.hist([sugar_level_men,sugar_level_women],bins=[80,100,125,200],rwidth=0.97,label=['men','women'],color=['red','black'],orientation='horizontal')
plt.legend(fontsize='large')
```

# Pie chart

```python
house_expenses=[6000,3000,2000,1500,14000,500]
expense_names=['Electricity','Internet','School Fee','Van','House Rent','Water']
```

```python
plt.pie(house_expenses,labels=expense_names,radius=2,shadow=True,autopct='%0.0f%%',
        colors=['grey','yellow','blue','green','pink','orange'],explode=[0.1,0.2,0.1,0,0,0],startangle=90)

plt.show()
```
