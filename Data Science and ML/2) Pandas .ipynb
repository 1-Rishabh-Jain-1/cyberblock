# Different ways of creating a dataframe

```python
import pandas as pd
import numpy as np
```

```python
#list of dictionaries to a dataframe
weather_data=[{'date':'1/1/2002','temperature':32,'Windspeed':10,'event':'Sunny' },
             {'date':'2/1/2002','temperature':25,'Windspeed':25,'event':'Rain' },
             {'date':'3/1/2002','temperature':16,'Windspeed':40,'event':'Cool' },
             {'date':'4/1/2002','temperature':45,'Windspeed':12,'event':'Hot' },
             {'date':'5/1/2002','temperature':22,'Windspeed':34,'event':'Rain' }]
df=pd.DataFrame(weather_data)
df
```

```python
#dictionary of lists to a dataframe
weather_data={ 'date':['1/1/2002','2/1/2002','3/1/2002','4/1/2002','5/1/2002'],
             'temperature':[32,25,16,45,22],
             'Windspeed':[10,25,40,12,34],
             'event':['Sunny','Rain','Cool','Hot','Rain']
             }
df=pd.DataFrame(weather_data)
df
```

```python
#list of tuples to a dataframe
weather_data=[('1/1/2002',32,10,'Sunny') ,
             ('2/1/2002',25,25,'Rain'),
             ('3/1/2002',16,40,'Cool'),
             ('4/1/2002',45,12,'Hot'),
             ('5/1/2002',22,34,'Rain')]
df=pd.DataFrame(weather_data,columns=['day','temperature','windspeed','event'])
df
```

```python
weather_data_2=[('1/2/2002',32,10,'Sunny') ,
             ('2/2/2002',25,25,'Rain'),
             ('3/2/2002',16,40,'Cool'),
             ('4/2/2002',45,12,'Hot'),
             ('5/2/2002',22,34,'Rain')]
df_2=pd.DataFrame(weather_data_2,columns=['day','temperature','windspeed','event'])
df_2
```

# Handling Missing values in a dataframe

```python
df =pd.read_csv('train.csv')
df
```

# fillna()

```python
#Replacing Null with 0 or any other number in a whole dataframe
df.fillna('0')
```

```python
#Replacing Null with 0 or any other number inside a specific column.
df['outcome']=df['outcome'].fillna('0')
```

```python
#By passing dictionary
df.fillna({"F1":0,
           "F2":0,
           "outcome":"No outcome"})
```

```python
df.fillna(method='ffill')
#ffill means forward fill. it means that the most previous value is copied to null entries
```

```python
df.fillna(method='bfill')
#bfill means backward fill. it means that the next value is copied to null entries.
```

```python
df.fillna(method='ffill',limit=2)
# limit=2 means that the previous value is copied to at most two consecutive null entries
```

# interpolate()

```python
df.interpolate()
```

```python
df.interpolate(method='time')
#Tere are many other methods
```

# dropna()

```python
#Drop all those rows in which any  entry is null
df.dropna()
```

```python
#Drop all those rows whose all entries are null
df.dropna(how='all') 
```

```python
# keep only those rows who contains >= 3 valid entries ( 3 not- NULL columns). all other are deleted.
df.dropna(thresh=2)
```

# goupyby()

```python
weather_data_bycities={ 'date':['1/1/2002','2/1/2002','3/1/2002','4/1/2002','5/1/2002','6/1/2002','7/1/2002','8/1/2002','9/1/2002','10/1/2002','11/1/2002','12/1/2002'],
              'city':['Lahore','Queta','Multan','Multan','Lahore','Karachi','Queta','Multan','Lahore','Queta','Karachi','Karachi'],
             'temperature':[32,25,16,45,22,42,3,51,27,18,48,37],
             'Windspeed':[10,25,40,12,34,70,43,84,101,65,39,56],
             'event':['Sunny','Rain','Sunny','Sunny','Cool','Cool','Rain','Sunny','Rain','Cool','Hot','Hot']
             }
df=pd.DataFrame(weather_data_bycities)
df
```

```python
g=df.groupby('city')
g
```

```python
for city,city_df in g:
    print(city)
    print(city_df)
```

```python
#To access a specific group
g.get_group('Lahore')
```

```python
#To find maximum temperature and wind speend in each city
g.max()
```

```python
#To find average temperature and wind speend in each city
g.mean()
```
